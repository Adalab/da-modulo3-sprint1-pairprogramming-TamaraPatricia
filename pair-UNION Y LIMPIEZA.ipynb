```python
import pandas as pd
import numpy as np
```

```python
pd1 = pd.read_csv('world-data-2023_part1.csv', index_col=0)
pd2 = pd.read_csv('world-data-2023_part2.csv', index_col=0)
pd.set_option('display.max_columns', None)
```

```python
pd1
```

```python
pd2
```

Utiliza el mÃ©todo de uniÃ³n de Pandas que consideres mÃ¡s adecuado para combinar los datos de ambos aÃ±os en un solo DataFrame.

```python
df = pd1.merge(pd2, left_on='Country', right_on='country')
df
```

Hemos unido las columnas utilizando 'merge' porque tienen una columna en comÃº y nombre distinto (merge nos permite especificar los distintos nombres)

DespuÃ©s de la uniÃ³n de datos, tenemos dos columnas de "country". Elimina una de ellas.

```python
df.drop('country', axis = 1, inplace = True)
```

Los nombres de las columnas no son homogeneos. Cambia los nombres de las columnas de tal forma que:

No tengan espacios.

EstÃ©n en minÃºscula.

No tengan parÃ©ntesis, es decir, quitar "(%)", "(Km2)".

Algunas columnas tiene "\n". Eliminalos de los nombres de las columnas.

Algunas columnas tienen ":". Eliminalos de los nombres de las columnas.

```python
nuevas_columnas = {columna: columna.lower().replace('(_%)','').replace(" ", "_").replace('(p/km2)','').replace('(%)','').replace('(km2)','').replace('\n','').replace(':','') for columna in df.columns}
df.rename (columns = nuevas_columnas, inplace = True)
df
```

La columnas coordinates tiene la latitud y la longitud en una sola columna. Crea dos columnas nuevas, una con la longitud y otra con la latitud. Una vez hecho, elimina la columna de coordinates.

```python
df['coordinates'] = df['coordinates'].str.replace("(", "").str.replace(")", "")
nuevas = df['coordinates'].str.split(',', expand= True)
df = pd.concat([df,nuevas],axis = 1)
```

```python
df.rename(columns={0: 'latitud', 1: 'longitud'}, inplace=True)
df.drop('coordinates', axis=1, inplace=True)
df
```

Las columnas unemployment_rate, total_tax_rate, tax_revenue, population_labor_force_participation, out_of_pocket_health_expenditure, gross_tertiary_education_enrollment, gross_primary_education_enrollment, forested_area, cpi_change, agricultural_land tienen "%". Elimina los "%" de los valores de las columnas.

```python
columnas= ['unemployment_rate','total_tax_rate','tax_revenue_','population_labor_force_participation_','out_of_pocket_health_expenditure','gross_tertiary_education_enrollment_','gross_primary_education_enrollment_','forested_area_','cpi_change_','agricultural_land(_%)']
for elemento in columnas:
    df[elemento] = df[elemento].str.replace('%','')
```

```python
df.head()
```

Haz lo mismo para las columnas de gasoline_price, gdp, minimum_wage, pero eliminando "$".

```python
columnas=['gasoline_price','gdp','minimum_wage']
for elemento in columnas:
    df[elemento] = df[elemento].str.replace('$','')
```
