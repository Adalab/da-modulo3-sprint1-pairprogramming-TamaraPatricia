3. Imputación de Valores Nulos: Implementa el método de imputación seleccionado en el paso 2 para llenar los valores nulos en las columnas.

```python
# COLUMNAS CATEGÓRICAS
lista_columnas =["abbreviation", "capital/major_city",'currency-code','largest_city','official_language','continent']

for col in lista_columnas:
    df[col] = df[col].fillna("Unknown")

df[lista_columnas].isnull().sum()
```

```python
# COLUMNAS NUMÉRICAS - DATOS NO SUSTITUIBLES
lista_columnas =["land_area", "armed_forces_size",'minimum_wage','calling_code','forested_area_','gdp','population','population_labor_force_participation_','urban_population','latitud','longitud','agricultural_land(_%)','cpi','gasoline_price','tax_revenue_','cpi_change_','gross_primary_education_enrollment_','gross_tertiary_education_enrollment_','out_of_pocket_health_expenditure','physicians_per_thousand','total_tax_rate','unemployment_rate']

for col in lista_columnas:
    df[col] = df[col].fillna("Unknown")

df[lista_columnas].isnull().sum()
```

```python
# COLUMNAS NUMÉRICAS - DATOS SUSTITUIBLES
imputer_iterative = IterativeImputer(max_iter = 20, random_state = 42)
# ajustamos y tranformamos los datos
imputer_iterative_imputado = imputer_iterative.fit_transform(df[['birth_rate','co2-emissions','fertility_rate','infant_mortality','life_expectancy','maternal_mortality_ratio']])
# comprobamos que nos devuelve un array
imputer_iterative_imputado
# añadimos la columna con los datos al df
df[[ "birth_rate_iterative", "co2-emissions_iterative","fertility_rate_iterative", "infant_mortality_iterative","life_expectancy_iterative", "maternal_mortality_ratio_iterative"]] = imputer_iterative_imputado
# comprobamos los nulos
df[["birth_rate_iterative", "co2-emissions_iterative","fertility_rate_iterative", "infant_mortality_iterative","life_expectancy_iterative", "maternal_mortality_ratio_iterative"]].isnull().sum()
```

```python
#instanciamos la clase del KNNImputer utilizando los 5 vecinos más cercanos
imputer_knn = KNNImputer(n_neighbors = 5)
# ajustamos y transformamos los datos
imputer_knn_imputado = imputer_knn.fit_transform(df[['birth_rate','co2-emissions','fertility_rate','infant_mortality','life_expectancy','maternal_mortality_ratio']])
# comprobamos que nos devuelve un array
imputer_knn_imputado
# añadimos la columna con los datos  al df
df[["birth_rate_knn", "co2-emissions_knn","fertility_rate_knn", "infant_mortality_knn","life_expectancy_knn", "maternal_mortality_ratio_knn"]] = imputer_knn_imputado
# comprobamos los nulos
df[["birth_rate_knn", "co2-emissions_knn","fertility_rate_knn", "infant_mortality_knn","life_expectancy_knn", "maternal_mortality_ratio_knn"]].isnull().sum()
```

Analizando el tipo de datos que contiene cada columna, decidimos:

- En las columnas categóricas los datos no son imputables, ya que se trata de nombres de ciudades, moneda, lenguaje oficial, continente...

- En las columnas numéricas, al tratarse de países cuyos datos no necesariamente están relacionados con los de el resto de países, hemos optado por sustituir los datos nulos por 'Unknown' salvo en las columnas donde figuran los ratios de natalidad, mortalidad, fertilidad, esperanza de vida...etc que creemos que se pueden estimar tomando como referencia los datos de sus países vecinos.

A pesar de que creemos que los datos calculados con el algortitmo KNN a priori van a dar mejor resultado, decidimos hacer la imputación de nulos tanto con KNN como con Imperative Imputer para poder compararlo.
